### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
# Guardamos en un diccionario para recorrer más fácil
datasets = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

# Exploración básica
for name, df in datasets.items():
    print(f"===== {name} =====")
    print("Dimensiones:", df.shape)
    print("\nColumnas:", df.columns.tolist())
    print("\nTipos de datos:")
    print(df.dtypes)
    print("\nValores nulos:")
    print(df.isnull().sum())
    print("\nPrimeras filas:")
    display(df.head(3))
    print("="*60, "\n")
```

#1. Análisis de facturación



```python
ingresos = {
    "Tienda 1": tienda["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

for tienda_nombre, total in ingresos.items():
    print(f"{tienda_nombre}: ${total:,.2f}")
```

# 2. Ventas por categoría

```python
# Agrupar y contar ventas por categoría en cada tienda
categorias_t1 = tienda.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
categorias_t2 = tienda2.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
categorias_t3 = tienda3.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
categorias_t4 = tienda4.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)

print("===== Tienda 1 =====")
display(categorias_t1)

print("===== Tienda 2 =====")
display(categorias_t2)

print("===== Tienda 3 =====")
display(categorias_t3)

print("===== Tienda 4 =====")
display(categorias_t4)
```

# 3. Calificación promedio de la tienda


```python
valoraciones = {
    "Tienda 1": tienda["Calificación"].mean(),
    "Tienda 2": tienda2["Calificación"].mean(),
    "Tienda 3": tienda3["Calificación"].mean(),
    "Tienda 4": tienda4["Calificación"].mean()
}

for tienda_nombre, promedio in valoraciones.items():
    print(f"{tienda_nombre}: {promedio:.2f}")
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

def productos_populares(df, nombre_tienda):
    conteo = df.groupby("Producto")["Producto"].count().sort_values(ascending=False)

    print(f"===== {nombre_tienda} =====")
    print("\n📈 Top 5 productos más vendidos:")
    display(conteo.head(5))

    print("\n📉 Top 5 productos menos vendidos:")
    display(conteo.tail(5))

    # Visualización
    fig, axes = plt.subplots(1, 2, figsize=(14,5))

    conteo.head(5).plot(kind="bar", ax=axes[0], color="skyblue", title="Más vendidos")
    conteo.tail(5).plot(kind="bar", ax=axes[1], color="salmon", title="Menos vendidos")

    fig.suptitle(f"Productos más y menos vendidos - {nombre_tienda}", fontsize=14)
    plt.show()

# Aplicamos a cada tienda
productos_populares(tienda, "Tienda 1")
productos_populares(tienda2, "Tienda 2")
productos_populares(tienda3, "Tienda 3")
productos_populares(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
envio_promedio = {
    "Tienda 1": tienda["Costo de envío"].mean(),
    "Tienda 2": tienda2["Costo de envío"].mean(),
    "Tienda 3": tienda3["Costo de envío"].mean(),
    "Tienda 4": tienda4["Costo de envío"].mean()
}

for tienda_nombre, promedio in envio_promedio.items():
    print(f"{tienda_nombre}: ${promedio:,.2f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.bar(ingresos.keys(), ingresos.values(), color='skyblue')
plt.title("Ingreso total por tienda")
plt.ylabel("Ingresos ($)")
plt.xticks(rotation=0)
plt.show()
```

```python
import numpy as np

fig, ax1 = plt.subplots(figsize=(8,5))

tiendas = list(valoraciones.keys())
promedio_val = list(valoraciones.values())
promedio_envio = list(envio_promedio.values())

ax1.bar(tiendas, promedio_val, color='green', alpha=0.6, label='Calificación promedio')
ax1.set_ylabel('Calificación promedio', color='green')
ax1.set_ylim(0,5)

ax2 = ax1.twinx()
ax2.plot(tiendas, promedio_envio, color='red', marker='o', linewidth=2, label='Costo de envío promedio')
ax2.set_ylabel('Costo de envío ($)', color='red')

plt.title("Calificación y costo de envío promedio por tienda")
plt.show()
```

```python
import matplotlib.pyplot as plt

def mapa_ventas(df, nombre_tienda):
    plt.figure(figsize=(8,6))
    plt.scatter(df["lon"], df["lat"], alpha=0.6, c='blue', edgecolor='k', s=50)
    plt.title(f"Distribución geográfica de ventas - {nombre_tienda}")
    plt.xlabel("Longitud")
    plt.ylabel("Latitud")
    plt.grid(True)
    plt.show()

# Aplicamos a cada tienda
mapa_ventas(tienda, "Tienda 1")
mapa_ventas(tienda2, "Tienda 2")
mapa_ventas(tienda3, "Tienda 3")
mapa_ventas(tienda4, "Tienda 4")
```
